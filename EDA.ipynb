```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
df=df.drop(["Unnamed: 0"],axis=1)
df
```

```python
d={"cp":"chest pain type","trestbps":"resting blood pressure","chol":"serum cholestoral","fbs":"fasting blood sugar","restecg":"resting electrocardiographic results","thalach":"maximum heart rate achieved","exang":"exercise induced angina","oldpeak":"ST depression induced by exercise relative to rest","slope":"the slope of the peak exercise ST segment","ca":"number of major vessels (0-3) colored by flourosopy"}
```

```python
df.rename(columns=d,inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.svm import SVC
```

```python
X=df.drop("target",axis=1)
y=df.target
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=[LogisticRegression(),
       SVC(kernel="linear"),
       RandomForestClassifier(max_depth=5,n_estimators=150)]
```

```python
for m in model:
    model=m.fit(X_train,y_train)
    print("Model:",model)
    print("Training efficiency :",model.score(X_train,y_train))
    print("Testing Efficiency :",model.score(X_test,y_test))
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```
